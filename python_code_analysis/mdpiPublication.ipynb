```python
import numpy as np
import os

def is_dominated(p, candidates):
    for candidate in candidates:
        if all(candidate <= p) and any(candidate < p):
            return True
    return False

def find_pareto_front(points):
    pareto_front = []
    for point in points:
        if not is_dominated(point, points):
            pareto_front.append(point)
    return np.array(pareto_front)

def calculate_spread(pareto_front):
    if len(pareto_front) <= 1:
        return 0  # Spread is undefined or zero for a single point or empty set
    
    # Assuming sorting by the first objective for simplicity
    pareto_front_sorted = pareto_front[np.argsort(pareto_front[:, 0])]
    distances = np.sqrt(np.sum(np.diff(pareto_front_sorted, axis=0)**2, axis=1))
    d_bar = np.mean(distances)
    d_f = np.linalg.norm(pareto_front_sorted[0] - pareto_front_sorted[1])  # Distance to the first boundary
    d_l = np.linalg.norm(pareto_front_sorted[-1] - pareto_front_sorted[-2])  # Distance to the last boundary
    
    spread = (d_f + d_l + np.sum(np.abs(distances - d_bar))) / (d_f + d_l + (len(pareto_front) - 1) * d_bar)
    return spread



base_path = "/Users/adarshkesireddy/Downloads/NewData/Case_21"
directories = [ "nsgaii", "nsgaiii", "mpf", "empf"]

for directory in directories:
    print(f"Processing directory: {directory}")
    all_pareto_points = []

    for i in range(30):
        folder_path = os.path.join(base_path, str(i), directory, "rewards_999.txt")
        if os.path.isfile(folder_path):
            data = np.loadtxt(folder_path)
            pareto_front = find_pareto_front(data)
            all_pareto_points.extend(pareto_front)
        else:
            print(f"File not found: {folder_path}")

    if all_pareto_points:
        all_pareto_points = np.array(all_pareto_points)
        spread = calculate_spread(all_pareto_points)
        print(f"Spread metric for {directory}: {round(spread, 6)}")
    else:
        print("No data found for directory:", directory)
    print("-" * 80)
```

```python
import numpy as np
import os
import matplotlib.pyplot as plt
import pymoo

def is_dominated(p, candidates):
    for candidate in candidates:
        if np.all(np.array(candidate, dtype=float) <= np.array(p, dtype=float)) and \
           np.any(np.array(candidate, dtype=float) < np.array(p, dtype=float)):
            return True
    return False


def calculate_spacing(pareto_front):
    if len(pareto_front) < 2:
        return 0

    sorted_front = pareto_front[pareto_front[:, 0].argsort()]  # Sort for easier calculation
    distances = np.diff(sorted_front, axis=0) 
    return np.std(np.linalg.norm(distances, axis=1))

def find_pareto_front(points):
  pareto_front = []
  for point in points:
    if not is_dominated(point, points):
      pareto_front.append(point)
  return np.array(pareto_front)

def generational_distance(pareto_front, approximation):
    if len(pareto_front) == 0:
        return np.inf

    distances = np.linalg.norm(pareto_front[:, np.newaxis] - approximation, axis=2)
    min_distances = np.min(distances, axis=1)
    return np.mean(min_distances)

base_dir = "/Users/adarshkesireddy/Downloads/NewData/Case_21/"  # Base directory for simulations
optimizers = ["empf", "mpf","nsgaiii", "nsgaii"]

true_pf_filename = f"/Users/adarshkesireddy/Downloads/GD/test_function_21.txt"
true_pf = np.genfromtxt(true_pf_filename, delimiter="\t", usecols=(0, 1, 2), dtype=None)
true_pf = find_pareto_front(true_pf)
for optimizer in optimizers:
    all_generations = []
    for folder_num in range(30):
        filename = os.path.join(base_dir, str(folder_num), optimizer, "rewards_999.txt")
        data = np.genfromtxt(filename, delimiter="\t", usecols=(0, 1, 2), dtype=None)
        all_generations.append(data)

    combined_generation = np.concatenate(all_generations)
    pareto_front = find_pareto_front(combined_generation)

    # Calculate performance metrics
    gd = generational_distance(true_pf, pareto_front)
    spacing = calculate_spacing(pareto_front)

    print(f"Generational distance using {optimizer}:", round(gd,6))
    print(f"Spacing using {optimizer}:", spacing)
    



    
```
